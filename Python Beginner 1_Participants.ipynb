```python
# Reverse a Number Using while Loop

num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10  # Get the last digit
    reverse = reverse * 10 + digit
    num = num // 10  # Remove last digit

print("Reversed Number:", reverse)
```

```python
# Check if a String is a Palindrome

def is_palindrome(string):
    return string == string[::-1]

word = input("Enter a word: ")
if is_palindrome(word):
    print(word, "is a palindrome!")
else:
    print(word, "is NOT a palindrome.")
```

```python
def second_largest(numbers):
    unique_numbers = list(set(numbers))  # Remove duplicates
    unique_numbers.sort(reverse=True)  # Sort in descending order
    return unique_numbers[1] if len(unique_numbers) > 1 else None

nums = [10, 20, 30, 40, 50]
print("Second largest:", second_largest(nums))
```

Reverse a Dictionary

```python
# Swapping Keys and Values

def reverse_dict(d):
    return {v: k for k, v in d.items()}

original_dict = {"Alice": 25, "Bob": 30, "Charlie": 35}
reversed_dict = reverse_dict(original_dict)

print(reversed_dict)
```

### Types of Function Arguments

1ï¸âƒ£ Positional Arguments: Arguments are passed in order.


```python
def person(name, age):
    print(name, "is", age, "years old")

person("Alice", 25) # Correct
person(25, "Alice") # Incorrect order
```

2ï¸âƒ£ Keyword Arguments: Arguments are passed with variable names, ignoring order.

```python
def person(name, age):
    print(name, "is", age, "years old")

person(age=30, name="Bob")  # Correct order is not required
```

3ï¸âƒ£ Default Arguments

Default arguments are used when no value is passed.

If an argument is provided, it overrides the default value.

```python
# Function with Default Arguments

def greet(name="Guest"):
    print("Hello,", name)

greet()
greet("Alice")
```

```python
# Multiple Default Arguments

def order(item="Coffee", size="Medium"):
    print(f"Order: {size} {item}")

order()
order("Tea", "Large")
order(size="Small")
```

4ï¸âƒ£ Arbitrary Arguments (*args): Used when number of arguments is unknown.

*args allows passing a variable number of arguments.

```python
# Using *args for Variable Arguments

def add_numbers(*args):

    

add_numbers(1, 2, 3)
add_numbers(5, 10, 15, 20)
```

5ï¸âƒ£ Arbitrary Keyword Arguments (**kwargs): Used when passing multiple key-value arguments.

**kwargs allows passing keyword arguments (key-value pairs).

```python
# Using **kwargs for Keyword Arguments

def person_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

person_info(name="Alice", age=25, city="New York")
```

```python
# Using *args and **kwargs Together

def full_info(*args, **kwargs):
    print("Arguments:", args)
    print("Keyword Arguments:", kwargs)

full_info(1, 2, 3, name="Alice", age=30)
```

### Built-in Functions in Python

Python provides many built-in functions: predefined functions.

```python
# abs(), round(), pow()

print(abs(-10))
print(round(3.567, 2))
print(pow(2, 3))
```

```python
# map() Function
# The map() function applies a function to all elements in an iterable.

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
# filter() Function
# The filter() function filters elements based on a condition.

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

#### Lambda Functions

A lambda function is an anonymous function defined with the lambda keyword.

One-line function.

No def keyword needed.

```python
square = lambda x: x * x
print(square(5))
```

```python
# Lambda Function with map()

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
# Lambda Function with filter()

numbers = [10, 25, 30, 47, 50]
div_by_5 = list(filter(lambda x: x % 5 == 0, numbers))
print(div_by_5)
```

#### Recursive Functions

A recursive function calls itself until a base condition is met.

```python
# Factorial Using Recursion

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

```python
# Fibonacci Series Using Recursion

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(7):
    print(fibonacci(i), end=" ")
```

#### Nested Functions

A function defined inside another function.

```python
def outer():
    print("Outer function")

    def inner():
        print("Inner function")

    inner()

outer()
```

#### Function to Find Maximum of Three Numbers

```python
def max_of_three(a, b, c):
    return max(a, b, c)

print("Largest:", max_of_three(10, 25, 15))
```

```python
#  Alternative Without max() Function

def max_of_three(a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c

print("Largest:", max_of_three(10, 25, 15))
```

#### Function to Calculate the Factorial

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print("Factorial:", factorial(5))
```

#### Function to Check if a Number is Prime

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
```

#### Function to Generate Random Password

```python
import random
import string

def generate_password(length=8):
    characters = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choice(characters) for i in range(length))

print(generate_password(12))  # Output: Random 12-character password
```
